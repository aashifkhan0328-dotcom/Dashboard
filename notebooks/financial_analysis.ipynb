# Financial Performance & Risk Analysis Dashboard
## Walmart vs. Target - Comprehensive Financial Analysis

This notebook provides a detailed analysis of Walmart and Target's financial performance over the last 5 fiscal years (2020-2024).

**Objective:** Determine which company is financially stronger based on profitability, liquidity, solvency, and operational efficiency.

## 1. Import Required Libraries and Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings

warnings.filterwarnings('ignore')

# Set style for better visualizations
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (14, 6)

print("Libraries imported successfully!")
```

```python
# Load the combined financial data
df = pd.read_csv('data/cleaned/walmart_target_combined.csv')

# Load the financial ratios
df_ratios = pd.read_csv('data/cleaned/financial_ratios.csv')

print(f"Data shape: {df.shape}")
print(f"Ratios shape: {df_ratios.shape}")
print("\nFirst few rows of financial data:")
print(df.head())
```

## 2. Executive Summary - Key Performance Indicators

```python
# Get latest year data for each company
latest_walmart = df[df['company'] == 'Walmart'].iloc[0]
latest_target = df[df['company'] == 'Target'].iloc[0]

# Create comparison KPI table
kpi_comparison = pd.DataFrame({
    'Metric': ['Revenue ($M)', 'Net Income ($M)', 'Total Assets ($M)', 
               'Total Equity ($M)', 'Operating Cash Flow ($M)'],
    'Walmart': [
        f"${latest_walmart['revenue']:,.0f}",
        f"${latest_walmart['net_income']:,.0f}",
        f"${latest_walmart['total_assets']:,.0f}",
        f"${latest_walmart['total_equity']:,.0f}",
        f"${latest_walmart['operating_cash_flow']:,.0f}"
    ],
    'Target': [
        f"${latest_target['revenue']:,.0f}",
        f"${latest_target['net_income']:,.0f}",
        f"${latest_target['total_assets']:,.0f}",
        f"${latest_target['total_equity']:,.0f}",
        f"${latest_target['operating_cash_flow']:,.0f}"
    ]
})

print("\n" + "="*70)
print("EXECUTIVE SUMMARY - KEY PERFORMANCE INDICATORS (FY 2024)")
print("="*70)
print(kpi_comparison.to_string(index=False))
print("="*70)
```

```python
# Create interactive KPI comparison visualization
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=("Revenue Comparison", "Net Income Comparison",
                    "Total Assets Comparison", "Operating Cash Flow Comparison"),
    specs=[[{}, {}], [{}, {}]]
)

# Revenue
fig.add_trace(
    go.Bar(x=['Walmart', 'Target'], y=[latest_walmart['revenue'], latest_target['revenue']],
           name='Revenue', marker_color='#0071ce'),
    row=1, col=1
)

# Net Income
fig.add_trace(
    go.Bar(x=['Walmart', 'Target'], y=[latest_walmart['net_income'], latest_target['net_income']],
           name='Net Income', marker_color='#cc0000'),
    row=1, col=2
)

# Total Assets
fig.add_trace(
    go.Bar(x=['Walmart', 'Target'], y=[latest_walmart['total_assets'], latest_target['total_assets']],
           name='Total Assets', marker_color='#4CAF50'),
    row=2, col=1
)

# Operating Cash Flow
fig.add_trace(
    go.Bar(x=['Walmart', 'Target'], y=[latest_walmart['operating_cash_flow'], latest_target['operating_cash_flow']],
           name='Operating Cash Flow', marker_color='#FF9800'),
    row=2, col=2
)

fig.update_layout(height=800, title_text="Executive Summary - Key Metrics (FY 2024)", showlegend=False)
fig.show()
```

## 3. Profitability Analysis

```python
# Revenue and Net Income Trends
fig = go.Figure()

walmart_data = df[df['company'] == 'Walmart'].sort_values('year')
target_data = df[df['company'] == 'Target'].sort_values('year')

fig.add_trace(go.Scatter(x=walmart_data['year'], y=walmart_data['revenue'],
                         mode='lines+markers', name='Walmart Revenue', line=dict(color='#0071ce', width=3)))
fig.add_trace(go.Scatter(x=target_data['year'], y=target_data['revenue'],
                         mode='lines+markers', name='Target Revenue', line=dict(color='#cc0000', width=3)))

fig.update_layout(title='Revenue Trend (FY 2020-2024)',
                  xaxis_title='Fiscal Year',
                  yaxis_title='Revenue ($M)',
                  hovermode='x unified',
                  height=500)
fig.show()

print("\nRevenue Analysis:")
print(f"Walmart Revenue Growth (2020-2024): {((walmart_data.iloc[-1]['revenue'] - walmart_data.iloc[0]['revenue']) / walmart_data.iloc[0]['revenue'] * 100):.1f}%")
print(f"Target Revenue Growth (2020-2024): {((target_data.iloc[-1]['revenue'] - target_data.iloc[0]['revenue']) / target_data.iloc[0]['revenue'] * 100):.1f}%")
```

```python
# Profitability Ratios Comparison
profitability_metrics = ['Gross_Margin_%', 'Net_Margin_%', 'ROE_%']

fig = go.Figure()

# Add Walmart traces
for metric in profitability_metrics:
    walmart_data_sorted = walmart_data.sort_values('year')
    fig.add_trace(go.Scatter(x=walmart_data_sorted['year'], y=walmart_data_sorted[metric],
                            mode='lines+markers', name=f'Walmart {metric}',
                            line=dict(dash='solid')))

# Add Target traces
for metric in profitability_metrics:
    target_data_sorted = target_data.sort_values('year')
    fig.add_trace(go.Scatter(x=target_data_sorted['year'], y=target_data_sorted[metric],
                            mode='lines+markers', name=f'Target {metric}',
                            line=dict(dash='dash')))

fig.update_layout(title='Profitability Ratios Trend',
                  xaxis_title='Fiscal Year',
                  yaxis_title='Percentage (%)',
                  hovermode='x unified',
                  height=600)
fig.show()
```

## 4. Liquidity, Solvency, and Risk Analysis

```python
# Current Ratio and Debt-to-Equity Trends
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=("Current Ratio Trend", "Debt-to-Equity Ratio Trend")
)

# Current Ratio
walmart_sorted = walmart_data.sort_values('year')
target_sorted = target_data.sort_values('year')

fig.add_trace(
    go.Scatter(x=walmart_sorted['year'], y=walmart_sorted['Current_Ratio'],
               mode='lines+markers', name='Walmart', line=dict(color='#0071ce')),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(x=target_sorted['year'], y=target_sorted['Current_Ratio'],
               mode='lines+markers', name='Target', line=dict(color='#cc0000')),
    row=1, col=1
)

# Debt-to-Equity
fig.add_trace(
    go.Scatter(x=walmart_sorted['year'], y=walmart_sorted['Debt_to_Equity'],
               mode='lines+markers', name='Walmart', line=dict(color='#0071ce'), showlegend=False),
    row=1, col=2
)

fig.add_trace(
    go.Scatter(x=target_sorted['year'], y=target_sorted['Debt_to_Equity'],
               mode='lines+markers', name='Target', line=dict(color='#cc0000'), showlegend=False),
    row=1, col=2
)

fig.update_yaxes(title_text="Current Ratio", row=1, col=1)
fig.update_yaxes(title_text="Debt-to-Equity", row=1, col=2)
fig.update_xaxes(title_text="Fiscal Year", row=1, col=1)
fig.update_xaxes(title_text="Fiscal Year", row=1, col=2)

fig.update_layout(height=500, hovermode='x unified')
fig.show()

print("\nLiquidity & Solvency Analysis (FY 2024):")
print(f"\nWalmart:")
print(f"  Current Ratio: {latest_walmart['Current_Ratio']:.2f}")
print(f"  Debt-to-Equity: {latest_walmart['Debt_to_Equity']:.2f}")
print(f"\nTarget:")
print(f"  Current Ratio: {latest_target['Current_Ratio']:.2f}")
print(f"  Debt-to-Equity: {latest_target['Debt_to_Equity']:.2f}")
```

## 5. Efficiency and Market Performance

```python
# Efficiency Ratios Comparison
efficiency_metrics = ['Inventory_Turnover', 'Asset_Turnover', 'Receivables_Turnover']

latest_walmart_ratios = df_ratios[df_ratios['company'] == 'Walmart'].iloc[0]
latest_target_ratios = df_ratios[df_ratios['company'] == 'Target'].iloc[0]

fig = go.Figure(data=[
    go.Bar(x=efficiency_metrics, 
           y=[latest_walmart_ratios['Inventory_Turnover'], 
              latest_walmart_ratios['Asset_Turnover'],
              latest_walmart_ratios['Receivables_Turnover']], 
           name='Walmart', marker_color='#0071ce'),
    go.Bar(x=efficiency_metrics, 
           y=[latest_target_ratios['Inventory_Turnover'], 
              latest_target_ratios['Asset_Turnover'],
              latest_target_ratios['Receivables_Turnover']], 
           name='Target', marker_color='#cc0000')
])

fig.update_layout(title='Efficiency Ratios Comparison (FY 2024)',
                  xaxis_title='Ratio Type',
                  yaxis_title='Ratio Value',
                  barmode='group',
                  height=500)
fig.show()
```

## 6. Cash Flow Analysis

```python
# Operating Cash Flow Trend
fig = go.Figure()

walmart_sorted = walmart_data.sort_values('year')
target_sorted = target_data.sort_values('year')

fig.add_trace(go.Bar(x=walmart_sorted['year'], y=walmart_sorted['operating_cash_flow'],
                    name='Walmart', marker_color='#0071ce'))
fig.add_trace(go.Bar(x=target_sorted['year'], y=target_sorted['operating_cash_flow'],
                    name='Target', marker_color='#cc0000'))

fig.update_layout(title='Operating Cash Flow Comparison (FY 2020-2024)',
                  xaxis_title='Fiscal Year',
                  yaxis_title='Operating Cash Flow ($M)',
                  barmode='group',
                  height=500)
fig.show()

print(f"\nCash Flow Analysis (FY 2024):")
print(f"\nWalmart:")
print(f"  Operating Cash Flow: ${latest_walmart['operating_cash_flow']:,.0f}M")
print(f"  Free Cash Flow: ${latest_walmart['free_cash_flow']:,.0f}M")
print(f"  OCF Margin: {latest_walmart_ratios['OCF_Margin_%']:.2f}%")
print(f"\nTarget:")
print(f"  Operating Cash Flow: ${latest_target['operating_cash_flow']:,.0f}M")
print(f"  Free Cash Flow: ${latest_target['free_cash_flow']:,.0f}M")
print(f"  OCF Margin: {latest_target_ratios['OCF_Margin_%']:.2f}%")
```

## 7. Comprehensive Financial Ratio Summary

```python
# Create comprehensive ratio comparison table
ratio_columns = ['Gross_Margin_%', 'Operating_Margin_%', 'Net_Margin_%', 'ROA_%', 'ROE_%',
                 'Current_Ratio', 'Quick_Ratio', 'Debt_to_Equity', 'Debt_Ratio_%',
                 'Inventory_Turnover', 'Asset_Turnover', 'OCF_Margin_%']

latest_walmart_all = df_ratios[df_ratios['company'] == 'Walmart'].iloc[0]
latest_target_all = df_ratios[df_ratios['company'] == 'Target'].iloc[0]

ratio_summary = pd.DataFrame({
    'Ratio': ratio_columns,
    'Walmart': [latest_walmart_all[col] for col in ratio_columns],
    'Target': [latest_target_all[col] for col in ratio_columns]
})

print("\n" + "="*80)
print("COMPREHENSIVE FINANCIAL RATIO ANALYSIS (FY 2024)")
print("="*80)
print(ratio_summary.to_string(index=False))
print("="*80)
```
